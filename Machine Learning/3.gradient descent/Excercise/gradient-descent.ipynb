```python
import pandas as pd
import numpy as np
```

```python
def gradient_descent(x,y):
    m=b=0
    n=len(x)
    iterations=100000
    lr=0.0001
    for i in range(iterations):
        y_hat=m*x+b
        dy_hat__dm=(-2/n)*sum(x*(y-y_hat))
        dy_hat__db=(-2/n)*sum(y-y_hat)
        m=m-lr*dy_hat__dm
        b=b-lr*dy_hat__db
        loss=(1/n)*sum([val**2 for val in (y-y_hat)])
        print("m :{},b:{} at iteration:{}, loss:{}".format(m,b,i,loss))
        
        
    return m,b
```

```python
df=pd.read_csv('test_scores.csv')
df
```

```python
x= df.math
y=df.cs
```

```python
x=np.array(x)
y=np.array(y)
```

```python
gradient_descent(x,y)
```
