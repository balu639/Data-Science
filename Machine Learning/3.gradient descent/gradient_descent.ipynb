Implementing gradient descent


```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m=b=0
    iterations=10000
    n= len(x)
    lr= 0.08
    for i in range(iterations):
        y_hat=m*x+b
        dy_hat__dm= -(2/n)*sum(x*(y-y_hat))
        dy_hat__db= -(2/n)*sum(y-y_hat)
        m=m-lr*dy_hat__dm
        b=b-lr*dy_hat__db
        loss= 1/n*sum([val**2 for val in (y-y_hat)])
        print("m {},b {}, iteration {},loss {} at {}".format(m,b,i,loss,i))
        
    return m,b
        
```

```python
x= np.array([1,2,3,4,5])
y= np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
