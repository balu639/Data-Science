```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv("canada-income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income in USD")
plt.scatter(df.year,df.income, color="red", marker="*")
```

```python
regression= linear_model.LinearRegression()
regression.fit(df[['year']],df.income)

```

```python
regression.predict([[2016]])
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income in USD")
plt.scatter(df.year,df.income, color="red", marker="*")
plt.plot(df.year, regression.predict(df[['year']]), color= "blue")
```
