```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
df_dummies = pd.get_dummies(df['Car Model'])
df_dummies
```

```python
final_df = pd.concat([df,df_dummies], axis = 'columns')
final_df
```

```python
final_df = final_df.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final_df
```

```python
X = final_df.drop(['Sell Price($)'], axis = 'columns')
X
```

```python
y = final_df['Sell Price($)']
y
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.predict([[45000,4,0,0]])
```

```python
reg.predict([[86000,7,0,1]])
```

```python
reg.score(X,y)
```
