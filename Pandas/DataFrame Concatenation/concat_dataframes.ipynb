```python
import pandas as pd
```

```python
weather_india = pd.DataFrame({
    'city': ['mumbai', 'kolkata', 'chennai'],
    'temperature': [34,24,32],
    'windspeed': [5,7,12]
})

weather_india
```

```python
weather_usa = pd.DataFrame({
    'city': ['new york', 'Indianapolis', 'chicago'],
    'temperature': [21,4,12],
    'windspeed': [12,23,32]
})

weather_usa
```

```python
new_df = pd.concat([weather_usa, weather_india])

new_df #Output will have the indexes from the original data frames
```

```python
new_df = pd.concat([weather_usa, weather_india], ignore_index= True) #Ignores the indexes from the original dataframes


new_df
```

```python
new_df = pd.concat([weather_usa, weather_india], keys=['usa', 'india']) #combining the two data frames using the primary keys

new_df
```

```python
new_df.loc['usa']
```

```python
new_df.loc['india']
```

```python
temperatures_df = pd.DataFrame({
    'city': ['Mumbai', 'Chennai', 'Kolkata'],
    'temperature': [32,34,23]
}, index=[0,1,2])

temperatures_df
```

```python
windspeeds_df = pd.DataFrame({
    'city': ['Chennai','Mumbai','Kolkata'],
    'windspeed': [12,15,18]
}, index=[1,0,2])

windspeeds_df
```

```python
pd.concat([temperatures_df,windspeeds_df], axis = 1)
```

```python
#concatenating a pandas series to a dataframe
event = pd.Series(['Rainy', 'Humid', 'Windy'], name='event')
event
```

```python
df = pd.concat([temperatures_df,event], axis = 1)

df
```
