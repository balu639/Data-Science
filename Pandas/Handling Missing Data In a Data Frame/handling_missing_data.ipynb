```python
df = pd.read_csv('weather_data.csv', parse_dates = ['day']) #reading the data frame by parsing the day column as datetime

type(df['day'][0])
```

```python
#setting the index to day column instead the automated index column from 0 to 8

df.set_index('day', inplace=True)
```

```python
df
```

```python
#fillna: Replace all the NaN values in the above data frame to a specified value

new_df = df.fillna(0)
new_df
```

```python
new_df1 = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No event'
})

new_df1
```

```python
new_df2 = df.fillna(method='ffill') # method = 'ffill' fill na values with the value from previous row or column

new_df2
```

```python
new_df3 = df.fillna(method = 'bfill', axis = 'rows') # method = 'bfill' fills na values with the value from the next row or column

new_df3
```

```python
new_df4 = df.fillna(method = 'bfill', axis = 'columns')

new_df4
```

```python
new_df5 = df.fillna(method ='ffill', limit= 1) #limiting how many rows or columns it should copy the same value from previous row or column

new_df5
```

```python
#Instead of just copying the values from previous or after rows/columns for NaN, Interpolate generates the values according to the data given 

new_df6 = df.interpolate()

new_df6
```

```python
#Interpolate using the time or day: It'll fill the NaN values based on the date time trend

new_df7 = df.interpolate(method='time')

new_df7
```

```python
#Dropping the data when there is a NaN values in any of it's column

new_df8 = df.dropna()

new_df8
```

```python
#Dropping the data column when there is NaN in all of it's columns

new_df9 = df.dropna(how='all')

new_df9
```

```python
#adding the missed dates to the above time series index
dt = pd.date_range('01-01-2017','01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
