```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('weather_data 2.csv')
df
```

```python
#replace the special value(-99999) with NaN

new_df = df.replace(-99999, np.NaN)

new_df
```

```python
new_df= df.replace({
    'temperature': -99999,
    'windspeed': -88888,
    'event': '0'
}, np.NaN)

new_df
```

```python
new_df = df.replace({
    -99999 : np.NaN,
    '0': 'Sunny'
    
})

new_df
```

```python
#Regular expressions to remove units of measures
df = pd.read_csv('weather_data 2.csv')

df
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
    
}, '', regex=True)

new_df
```
