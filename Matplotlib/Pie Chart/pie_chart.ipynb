```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
values= [353,55,30,150,50]
value_labels = ['House Rent','Internet','Commute','Groceries','Miscellaneous']
```

```python
plt.pie(values, labels=value_labels, radius= 1.6,autopct = '%0.0f%%')
plt.show()
```

```python
plt.pie(values, labels=value_labels, radius= 1.6,autopct = '%0.2f%%', shadow = True)
plt.show()
```

```python
#Seperate the pies from the pie chart for visual representation

plt.pie(values, labels=value_labels, radius= 1.6,autopct = '%0.2f%%', explode = [0,0.5,0.2,0,0])
plt.show()
```

```python
#rotate the pie chart using startangle arguement

plt.pie(values, labels=value_labels, radius= 1.6,autopct = '%0.2f%%', explode = [0,0.5,0.2,0,0], startangle=45)
plt.show()
```
