```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL', "AMAZON", "MICROSOFT", "META"]
revenue = [94,120,123,82]
profit = [20,40,32,10]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel('Tech Comapanies')
plt.ylabel('Revenue(Billn)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4, label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4, label='Profit')
plt.legend()
```

```python
#Horizontal bar charts

plt.yticks(xpos,company)
plt.ylabel('Tech Comapanies')
plt.xlabel('Revenue(Billn)')
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue,height=0.4, label='Revenue')
plt.barh(xpos+0.2,profit,height=0.4, label='Profit')
plt.legend()
```
